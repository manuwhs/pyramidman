# Currencies 

Gathering of the Forex functionalities of the library.

### Import libraries

```python
import sys
import datetime as dt
import pandas as pd
import numpy as np
from IPython.display import Image
from pandas.plotting import scatter_matrix
%matplotlib qt 

sys.path.insert(0,"..") # Adds higher directory to python modules path.

from traphing.data_classes import Velas, Symbol, Portfolio
from traphing.utils import Timeframes, unwrap
import traphing.utils as ul
from traphing.graph.Gl import gl
```

```python
portfolio_id = "Manu's winter portfolio"
symbol_names = []
timeframes = [Timeframes.M15, Timeframes.D1]

dataSource =  "MQL5"  # Hanseatic  FxPro GCI Yahoo
[storage_folder, updates_folder] = ul.get_foldersData(source = dataSource, rrf = "../../../" )
folder_images = "./images/python_generated/currencies/"
```

```python
portfolio = Portfolio(portfolio_id, symbol_names, timeframes)
symbols_properties_df = Symbol.load_symbols_properties_csv(storage_folder)
all_symbol_names = symbols_properties_df.index.to_list()

## Empty portfolio for getting the desired available exchanges
all_portfolio = Portfolio("all currencies", all_symbol_names, [])
```

## Related exchanges

A currency is in several exchanges.

```python
currency = "CHF"
symbol_names = all_portfolio.get_related_forex_currencies(currency)
print (symbol_names)
```

```python
portfolio = Portfolio("CHF related currencies", symbol_names, timeframes)
portfolio.load_data_from_csv(storage_folder)
```

```python
start_time = dt.datetime(2019,7,20); end_time = dt.datetime(2019,7,25)
portfolio.set_time_interval(start_time,end_time)
```

```python
currencies_df = portfolio.map_timeframe(timeframe = Timeframes.M15, method_name = "series", name = "Close")
currencies_df
```

```python
image_name = "raw_currencies.png"; img_path = folder_images + image_name
size_inches = [12, 8]; 
gl.init_figure()
nrows = len(symbol_names) 

i = 0; axes = None
for symbol_name in symbol_names:
    axes = gl.subplot2grid((nrows,1),(i,0),sharex = axes)
    gl.plot(currencies_df.index, currencies_df[symbol_name], axes = axes, labels = ["","",symbol_name])
    i+=1 
gl.subplots_adjust(left=.09, bottom=.10, right=.90, top=.95, wspace=.20, hspace=0, hide_xaxis = True)
gl.save_figure(img_path, dpi = 100, size_inches = size_inches, close = True)
Image(img_path)
```

```python
currencies_returns_df = currencies_df.pct_change()
```

```python
image_name = "raw_currencies_returns.png"; img_path = folder_images + image_name
size_inches = [12, 8]; 
gl.init_figure()
nrows = len(symbol_names) 

i = 0; axes = None
for symbol_name in symbol_names:
    axes = gl.subplot2grid((nrows,1),(i,0),sharex = axes)
    gl.stem(currencies_returns_df.index, currencies_returns_df[symbol_name], axes = axes, labels = ["","",symbol_name])
    i+=1 
gl.subplots_adjust(left=.09, bottom=.10, right=.90, top=.95, wspace=.20, hspace=0, hide_xaxis = True)
gl.save_figure(img_path, dpi = 100, size_inches = size_inches, close = True)
Image(img_path)
```

## Plot relationships

We can plot the values of each currency pair against each other. As we can appreciate, their values are correlated, 

```python
image_name = "currencies_relations.png"; img_path = folder_images + image_name
gl.init_figure()
axes = gl.subplot2grid((1,1),(0,0))
scatter_matrix(currencies_returns_df[currencies_returns_df.columns[:4]], ax = axes)

gl.save_figure(img_path, dpi = 100, size_inches = size_inches, close = True)
Image(img_path)
```

# Currency cycle

```python
currencies = ["EUR", "CHF", "USD"]
all_symbol_names = list(symbols_properties_df.index) 
exchanges =  ul.get_exchange_symbol_names(currencies, all_symbol_names)
print(exchanges)
```

```python
for exchange in exchanges:
    portfolio.add_symbol(exchange, timeframes)
    portfolio[exchange].load_data_from_csv(storage_folder)
```

```python
start_time = dt.datetime(2019,7,9); end_time = dt.datetime(2019,7,25)
portfolio.set_time_interval(start_time,end_time)
```

```python
currencies_df = portfolio.map_timeframe(timeframe = Timeframes.M15, method_name = "series", name = "Close")
```

## Visualize the cycle

We can create nodes and edges in the networkx library and plot it easily.
The values of the graph are logarithmic and should be 0.


```python
G = ul.get_forex_graph(currencies,currencies_df,currencies_df.index[0] )

image_name = "cycle_graph.png"; img_path = folder_images + image_name
size_inches = [7, 3]; 
gl.init_figure()
axes = gl.subplot2grid((1,1),(0,0))
ul.draw_graph(G, ax = axes)
gl.save_figure(img_path, dpi = 100, size_inches = size_inches, close = True)
Image(img_path)
```

```python
forex_cycle = portfolio.indicator(timeframe = Timeframes.M15, 
                                  name = "forex_cycle", currencies = ["EUR", "CHF", "USD"])
print (forex_cycle)
```

```python
folder_images = "./images/python_generated/Portfolio/"
image_name = "forex_cycle.png"; img_path = folder_images + image_name

gl.init_figure()
n_rows, n_cols = 5,1; size_inches = [12, 10]
i = 2; axes = None

for symbol_name in exchanges:
    axes = gl.subplot2grid((nrows,1),(i,0),sharex = axes)
    gl.plot(currencies_df.index, currencies_df[symbol_name], axes = axes, labels = ["","",symbol_name])
    i+=1 
    
axes = gl.subplot2grid((nrows,1),(0,0),sharex = axes, rowspan = 2)

gl.plot(forex_cycle.index, forex_cycle, 
        legend = [forex_cycle.name], labels = ["Forex cycle","",forex_cycle.name])

gl.subplots_adjust(left=.09, bottom=.10, right=.90, top=.95, wspace=.20, hspace=0, hide_xaxis = True)
gl.save_figure(img_path, dpi = 100, size_inches = size_inches, close = True)
Image(img_path)
```
