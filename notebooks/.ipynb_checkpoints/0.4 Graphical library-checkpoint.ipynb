## Velas plotting 



```python
symbol_name = "AUDCHF"
timeframe = Timeframes.M15
storage_folder = "../tests/data/storage/"
start_time = dt.datetime(2019,7,20)
end_time = dt.datetime(2019,8,20)

my_velas_M15 = Velas(symbol_name, timeframe)
my_velas_M15.load_data_from_csv(storage_folder)
my_velas_M15.set_time_interval(start_time, end_time, trim = False)

timestamps_M15 = my_velas_M15.timestamps
close_values_M15 = my_velas_M15["Close"]
```

As matplotlib internally only works with numeric values, to create advanced plotting with datetimes, we need to first converrsion.

For the conversion matplotlib provides matplotlib.dates.date2num. This understands datetime objects, so you would first need to convert your time series to datetime using series.index.to_pydatetime() and then apply date2num.


You can then work with the numeric points as usual , e.g. plotting as Polygon or LineCollection[1,2].

```python
inxval = mdates.date2num(timestamps_M15.to_pydatetime())
inxval
```

That is what the utilities to_mdates is for)
